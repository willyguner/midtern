```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, FloatSlider
```

```python
from keras.models import Sequential,Model
from keras.layers import Input,Lambda
from keras.layers import Dense #NN
from keras.layers import Conv2D,MaxPooling2D #CNN
from keras.optimizers import SGD,Adam
from keras import metrics
from keras.datasets import mnist
from keras.utils import np_utils
from keras import backend as k
```

```python
(x_train0,y_train0),(x_test0,y_test0)=mnist.load_data()
```

```python
x_train=x_train0.reshape(60000,-1) #-1為numpy自動計算
x_test=x_test0.reshape(10000,-1) #-1為numpy自動計算

x_train-=x_train.min()
x_train=x_train/x_train.max()

x_test-=x_test.min()
x_test=x_test/x_test.max()
```

```python
enc_1=Dense(100,activation='relu')

enc_mean=Dense(2)
enc_log_var=Dense(2)

dec_2=Dense(100,activation='relu')
dec_1=Dense(784,activation='sigmoid')
```

```python
x=Input(shape=(784,))

enc_x=enc_1(x)

z_mean=enc_mean(enc_x)
z_log_var=enc_log_var(enc_x)
```

```python
def sampling(args):
    z_mean,z_log_var=args
    epsilon=k.random_normal(shape=(2,),mean=0,stddev=1) #X∼N(0,1)
    return z_mean+k.exp(z_log_var/2)*epsilon
    
```

```python
z=Lambda(sampling,output_shape=(2,))([z_mean,z_log_var])
```

```python
x_hat=dec_1(dec_2(z))
```

```python
VAE=Model(x,x_hat)
VAE.summary()
```

```python
def vae_loss(x,x_recon):
    recovery_loss=784*metrics.binary_crossentropy(x,x_recon)
    KL_loss=-0.5*k.sum(1+z_log_var-k.square(z_mean)-k.exp(z_log_var),axis=-1)
    return recovery_loss+KL_loss
```

```python
VAE.compile(loss=vae_loss,optimizer=Adam())
VAE.fit(x_train,x_train,batch_size=64,epochs=20)
```

```python
VAE_encoder=Model(x,z_mean)
VAE_encoder.summary()
```

```python
h_input=Input(shape=(2,))
VAE_decoder=Model(h_input,dec_1(dec_2(h_input)))
VAE_decoder.summary()
```

```python
inds=np.random.randint(10000,size=3000)
VAE_latent=VAE_encoder.predict(x_train[inds])
VAE_latent.shape
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(VAE_latent[:,0],VAE_latent[:,1],c=y_train0[inds],cmap='tab10')
plt.colorbar()
```

```python
def normalized(x):
    x-=x.min()
    x/=x.max()
    return x
```

```python
n=15
digit_size=28
grid_x_vae = np.linspace(-4+0.05, 4-0.05, n)
grid_y_vae = np.linspace(-4+0.05, 4-0.05, n)
VAE_figure = np.zeros((digit_size * n, digit_size * n))
for i, yi in enumerate(grid_x_vae):
    for j, xi in enumerate(grid_y_vae):
        z_sample = np.array([[xi, yi]])
        x_decoded = VAE_decoder.predict(z_sample)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        VAE_figure[(n-i-1) * digit_size: (n - i) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = normalized(digit)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(VAE_figure,cmap='Greys')
plt.show
```

```python
def inBetween(t):
    data_0 = x_train0[idx_1]
    data_1 = x_train0[idx_2]
    data_t = (1-t)*x_train0[idx_1] + t*x_train0[idx_2]

    mu_0 = VAE_encoder.predict(x_train[idx_1:idx_1+1]).squeeze()
    mu_1 = VAE_encoder.predict(x_train[idx_2:idx_2+1]).squeeze()
    mu_t = (1-t)*mu_0 + t*mu_1

    plt.figure(figsize=(10, 5))
    ax1 = plt.subplot(2, 1, 2)
    ax1.scatter(mu_0[0], mu_0[1])
    ax1.scatter(mu_1[0], mu_1[1])
    ax1.scatter(mu_t[0], mu_t[1])
    ax1.plot([mu_0[0], mu_1[0]], [mu_0[1], mu_1[1]])
    ax1.annotate(str(mu_t[0]) + ', ' + str(mu_t[1]) , (mu_t[0], mu_t[1]))
    

    ax2 = plt.subplot(2, 3, 1)
    ax2.imshow(data_0, cmap='Greys')

    ax3 = plt.subplot(2, 3, 2)
    ax3.imshow(data_t, cmap='Greys')

    ax4 = plt.subplot(2, 3, 3)
    ax4.imshow(data_1, cmap='Greys')
```

```python
idx_1, idx_2 = np.random.randint(x_test.shape[0], size=2)
interact(inBetween, t=FloatSlider(value=0,min=0,max=1.0,step=0.02))
```
