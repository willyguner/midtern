```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

# Loss Function

## $$x^5+2x^4-3x^3-4x^2-5x+6$$

## 微分後

## $$5x^4+8x^3-9x^2-8x-5$$

```python
x=np.arange(-3,2,0.001)
def L(x):
    return x**5+2*x**4-3*x**3-4*x**2-5*x+6
def dL(x):
    return 5*x**4+8*x**3-9*x**2-8*x-5
plt.plot(x,L(x))
```

```python
def GradientDescent(x):
    w=np.random.uniform(-2,2)
    lr=0.001
    for i in range(x):
        w-=lr*dL(w)
        print("w=%f"%w)
        print("loss=%f"%L(w))
        plt.scatter(w,L(w))
```

```python
GradientDescent(500)
plt.plot(x,L(x))
```

```python
GradientDescent(500)
plt.plot(x,L(x))
```

# 結論

透過每次疊代出來的值，我們可以確定GradientDescent確實往LossFunction的最低點移動
